## **HTTP Methods**

HTTP methods are the foundation of how clients and servers communicate. These methods define the type of action to be performed on a resource.

In FastAPI, you can define routes that respond to different HTTP methods. Here are some of the most commonly used HTTP methods:

### **1. GET  Can I see the menu?**

You're asking for something. You don't want to change anything, just read or view what's available.

Youre saying: Hey server, give me the list of items.
- No changes are made to the server.

**Example:** [get_example.py](get_example.py)





### **2. POST  I'd like to place an order.**

Youre sending something new to the server, like adding a new item to a database. You provide details (like food name, size, toppings), and the kitchen (server) prepares it.

Youre saying: Here's a new item to create.
- Needs a request body (like a JSON payload).

Example: [post_example.py](post_example.py)



### **3. PUT  I want to change my whole order.**

You're replacing an existing resource with a new one. Like saying: "Forget the previous pizza. I want a completely new one"

Means: Replace item #1 with this entirely new version.
- You have to provide all fields again.
  
Example: [put_example.py](put_example.py)



### **4. DELETE  Cancel my order.**

Youre telling the server to remove something. Like canceling an order before its served.

Means: Remove item #1 from the system.

Example: [delete_example.py](delete_example.py)
